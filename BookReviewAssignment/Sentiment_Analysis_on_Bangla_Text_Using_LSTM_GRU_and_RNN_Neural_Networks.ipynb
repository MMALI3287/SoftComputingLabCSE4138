```python
stopdf = pd.read_csv('/content/stopwords_bangla.csv')
stopwords = stopdf['words'].tolist()
```

```python
def remove_stopwords(text):
  words = text.split()
  filtered_words = [word for word in words if word not in stopwords]
  return ' '.join(filtered_words)

df['review'] = df['review'].apply(remove_stopwords)
df
```

```python
from bnlp import BasicTokenizer

tokenizer = BasicTokenizer()

# raw_text = "আমি বাংলায় গান গাই"
# raw_text = df.iloc[0][1]
# tokens = tokenizer(raw_text)
# print(tokens)
# output: ["আমি", "বাংলায়", "গান", "গাই", "।"]
tokenizedData = []
for row in df.iterrows():
  raw_text = row[1][1]
  tokens = tokenizer(raw_text)
  tokenizedData.append(tokens)
print(tokenizedData[0])
```

```python
#max token size
max_token_size = 0
for tokens in tokenizedData:
  if len(tokens) > max_token_size:
    max_token_size = len(tokens)
print(max_token_size)
```

```python
from bnlp import BengaliWord2Vec

bwv = BengaliWord2Vec()

# word = 'গ্রাম'
# vector = bwv.get_word_vector(word)
# print(vector.shape)
# print(vector)
embeddedData = []
for tokens in tokenizedData:
  embedding = []
  for token in tokens:
    try:
      vector = bwv.get_word_vector(token)
      embedding.append(vector)
    except KeyError:
      continue
  embeddedData.append(embedding)
```

```python
from keras.preprocessing.sequence import pad_sequences
import numpy as np
embeddedData = pad_sequences(embeddedData, maxlen=50, padding='post', dtype='float32', value=np.zeros(100))
```

```python
embeddedData[0]
```

```python
# Hyperparameters
input_size = 100  #embedding dim
hidden_size = 100  #hidden dim
output_dim = 2
n_layers = 2
learning_rate = 0.0001
epochs_list = [5, 10, 30, 50, 70, 90, 110, 130, 155, 180]
batch_size = 32
```

```python
from torch.utils.data import Dataset, DataLoader
class CustomDataset(Dataset):
    def __init__(self, datas, labels):
        self.datas = np.array(datas)  # Convert Series to NumPy array
        self.labels = np.array(labels)  # Convert Series to NumPy array

    def __len__(self):
        return len(self.datas)

    def __getitem__(self, idx):
        text = self.datas[idx]
        label = self.labels[idx]
        return torch.tensor(text, dtype=torch.float32), torch.tensor(label, dtype=torch.long)
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
dataset = CustomDataset(embeddedData, df['rating'])
train_set , test_set = torch.utils.data.random_split(dataset, [int(0.8*len(dataset)), len(dataset) - int(0.8*len(dataset))])
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_set, batch_size=batch_size, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
class BiRNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(BiRNN, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True,bidirectional=True)
        self.fc = nn.Linear(hidden_size*2, num_classes)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size).to(device)
        out, _ = self.rnn(x, h0)
        #out: batch_size, seq_length,hidden_size
        out = out[:, -1, :]
        #out: batch_size, hidden_size
        out = self.fc(out)
        return out
```

```python
class BiGRU(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(BiGRU, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True,bidirectional=True)
        self.fc = nn.Linear(hidden_size*2, num_classes)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size).to(device)
        out, _ = self.gru(x, h0)
        #out: batch_size, seq_length,hidden_size
        out = out[:, -1, :]
        #out: batch_size, hidden_size
        out = self.fc(out)
        return out
```

```python
class BiLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(BiLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True,bidirectional=True)
        self.fc = nn.Linear(hidden_size*2, num_classes)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size).to(device)
        c0 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size).to(device)

        out, _ = self.lstm(x, (h0,c0))
        #out: batch_size, seq_length,hidden_size
        out = out[:, -1, :]
        #out: batch_size, hidden_size
        out = self.fc(out)
        return out
```

```python
import matplotlib.pyplot as plt
from tqdm import tqdm

def train_model(model, train_loader, test_loader, epochs_list, device, optimizer, criterion):
    all_losses = []  # List to store losses for each set of epochs
    evaluation_metrics = []  # List to store evaluation metrics after each set of epochs

    for epochs in epochs_list:
        model.train()
        epoch_losses = []  # List to store losses for the current set of epochs

        for epoch in range(epochs):
            total_loss = 0
            for i, (texts, labels) in enumerate(tqdm(train_loader)):
                texts = texts.to(device)
                labels = labels.to(device)

                # Forward pass
                outputs = model(texts)
                labels = labels.unsqueeze(1)
                loss = criterion(outputs, labels.long().squeeze(1))

                # Backward and optimize
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

                total_loss += loss.item()

            epoch_loss = total_loss / len(train_loader)
            epoch_losses.append(epoch_loss)
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {epoch_loss:.4f}')

        all_losses.append(epoch_losses)  # Append losses for current epochs to all_losses

        # Evaluate the model after current epochs
        accuracy, precision, recall, f1 = evaluate_model(model, test_loader, device)
        evaluation_metrics.append({
            'epochs': epochs,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1
        })

        # Reset the model for next set of epochs
        model.apply(lambda m: m.reset_parameters() if hasattr(m, 'reset_parameters') else None)  # Replace YourModelClass with your actual model instantiation

    return all_losses, evaluation_metrics
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
def evaluate_model(model, test_loader, device):
  model.eval()
  all_preds = []
  all_labels = []

  with torch.no_grad():
      for texts, labels in test_loader:
          texts = texts.to(device)
          labels = labels.to(device)
          predictions = model(texts)
          _, preds = torch.max(predictions, 1)
          all_preds.extend(preds.cpu().numpy())
          all_labels.extend(labels.cpu().numpy())

  accuracy = accuracy_score(all_labels, all_preds)
  precision = precision_score(all_labels, all_preds, average='binary')
  recall = recall_score(all_labels, all_preds, average='binary')
  f1 = f1_score(all_labels, all_preds, average='binary')

  print(f'Accuracy: {accuracy:.4f}')
  print(f'Precision: {precision:.4f}')
  print(f'Recall: {recall:.4f}')
  print(f'F1 Score: {f1:.4f}')
  return accuracy, precision, recall, f1
```

```python
# Initialize the model
model = BiRNN(input_size, hidden_size, n_layers, output_dim).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
print("BiRNN TRAINING")
rnnLoss,rnnmetrics = train_model(model, train_loader, test_loader, epochs_list, device, optimizer, criterion)
```

```python
# Initialize the model
model = BiGRU(input_size, hidden_size, n_layers, output_dim).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
print("BiGRU TRAINING")
gruLoss,gruMetrics = train_model(model, train_loader, test_loader, epochs_list, device, optimizer, criterion)
```

```python
# Initialize the model
model = BiLSTM(input_size, hidden_size, n_layers, output_dim).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
print("BiLSTM TRAINING")
lstmLoss,lstmMetrics = train_model(model, train_loader, test_loader, epochs_list, device, optimizer, criterion)
```

```python
epochs = [metric['epochs'] for metric in rnnmetrics]
accuracies = [metric['accuracy'] for metric in rnnmetrics]
precisions = [metric['precision'] for metric in rnnmetrics]
recalls = [metric['recall'] for metric in rnnmetrics]
f1_scores = [metric['f1'] for metric in rnnmetrics]

plt.plot(epochs, accuracies, label='Accuracy')
plt.plot(epochs, precisions, label='Precision')
plt.plot(epochs, recalls, label='Recall')
plt.plot(epochs, f1_scores, label='F1 Score')
plt.ylabel('Percentage')
plt.xlabel('Epoch')
plt.title('BiRNN Training Metrics')
plt.legend()
# plt.savefig('foo.png')
plt.show()
```
